```python
#@title Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
encoding = "utf-8"
```

```python
#@title  Load the csv file into a DataFrame
sms_data = pd.read_csv("/content/spam.csv", encoding='latin-1')
```

```python
sms_data.head(20)
```

```python
sms_data = sms_data[['v1', 'v2']]  # Select only the label (v1) and text (v2) columns
sms_data.columns = ['label', 'text']
```

```python
sms_data['label'] = sms_data['label'].map({'ham': 0, 'spam': 1})
```

```python
#@title train and text split
X_train, X_test, y_train, y_test = train_test_split(sms_data['text'], sms_data['label'], test_size=0.2, random_state=42)
```

```python
vectorizer = CountVectorizer(binary=True, stop_words='english')
X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)
```

```python
#@title Ml Algorithm
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_bow, y_train)
```

```python
#@title Making predictions
y_pred = nb_classifier.predict(X_test_bow)
```

```python
#@title Accuracy of the model built
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
